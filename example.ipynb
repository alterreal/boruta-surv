```python
import pandas as pd
from BorutaSurv import boruta_rsf, plot_feature_importances

SEED = 42 # Set random seed for reproduceable results
```

```python
# Get data (taken from https://dmkd.cs.vt.edu/projects/survival/data/)
data = pd.read_csv('lung.csv')

# Separate labels from features
y = data[['time', 'status']]
X = data.drop(columns=['time', 'status'])
```

```python
# Run boruta_rsf
important_features, tentative_features, importances, hits = boruta_rsf(X, y, trials = 10, random_state = SEED)

# Plot results
fig, ax = plot_feature_importances(X, importances, important_features, tentative_features)
```
